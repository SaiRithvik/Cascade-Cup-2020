```
import pylab as pl
data['Seasons'].hist(edgecolor='black', linewidth=1, figsize=(5, 5), grid=False,bins =4,rwidth=0.5)
pl.suptitle('Distribution of Season of absence')
```

```
data['Month of absence'].hist(edgecolor='black', linewidth=1.2, figsize=(5, 5), grid=False,bins=12)
pl.suptitle('Distribution of Month of absence')
```

```
data['Age'].hist(edgecolor='black', linewidth=1.2, figsize=(5, 5), grid=False,bins=10)
pl.suptitle('Distribution of Age of employees')
```

```
data['Education'].hist(edgecolor='black', linewidth=1.2, figsize=(5, 5), grid=False,bins=4, rwidth=0.5)
pl.suptitle('Distribution of Education of employees')
```

```
age_abs = data.groupby('Age')[['Absenteeism time in hours']].mean()
ax = age_abs.plot(kind='bar', figsize=(8,6), legend=False)
#for i, v in enumerate(age_abs.values):
#    ax.text(i-.25, v + 1, str(np.int(np.round(v))), color='red')
ax.set_ylabel('Absenteeism time in hours')
ax.set_title('Variation of Absenteeism time in hours by age')
plt.show()
```

```
month_abs = data.groupby('Month of absence')['Absenteeism time in hours'].sum()
ax = month_abs.plot(kind='bar', figsize=(8,6), legend=False)

ax.set_xlabel('Month')
ax.set_ylabel('Total Absentee hours')
ax.set_title('Total Absentee time in each month')
plt.show()
```
